```python
# Now the label encoding as we cannot pass the strings to our model while building it...
from sklearn.preprocessing import LabelEncoder
```

```python
cat_cols = ["workclass","education","marital-status","occupation","relationship","race","gender","native-country","income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(["income"],axis=1)
y = df["income"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Now since the data points are highly unbalanced i.e. some are really large values while some are really small, hence we will
# first scale our data & normalize it.
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
```

```python
#Fitting on the training data & tranforming both training & testing data
```

```python
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

# Optimization of the KNN model

### Optimizing K only...

```python
acc_lst = []
for i in range(1,35):
    model = KNeighborsClassifier(n_neighbors=i)
    model = model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc_lst.append(accuracy_score(y_test, y_pred))
```

```python
np.argmax(acc_lst)
```

```python
max(acc_lst)
```

```python
plt.figure(figsize=(15,8))
plt.plot(acc_lst)
plt.grid()
plt.show()
```
