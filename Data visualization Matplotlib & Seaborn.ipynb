```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import math
```

```python
sns.set()
```

```python
pie = math.pi
```

```python
pie
```

```python
x = np.linspace(0,2*pie,50)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y,color="r")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,color="b")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,color="b",marker="*")
plt.show()
```

```python
z = np.cos(x)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,z,c="g",marker="*",linestyle="--",label="Cos Curve")
plt.scatter(x,y,color="b",marker="^",s=100,label="Sine Curve")
plt.title("Sine & cosine visualization",fontsize=50)
plt.xlabel("radian values-->",fontsize=20)
plt.ylabel("Sin & cos values-->",fontsize=20)
plt.xticks(range(0,11))
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.show()
```
